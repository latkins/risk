```python
%matplotlib inline  
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
na = 1
nb = 1


initBench = 1
endBench = 10

initAsset = 1

mf = 1

sns.set(style="darkgrid", palette="Set2")
benchPure = np.linspace(initBench, endBench, 100)
benchNoise = np.random.normal(0,nb,100)
benchData = benchPure+benchNoise
r_b = (benchData[-1] - benchData[0]) / benchData[0]
sns.tsplot(benchData, color="blue")

alphas = []


def makeAssetData(v):
    endAsset = np.random.uniform(0,30)
                                
    assetPure = np.linspace(initAsset, endAsset, 100)
    assetNoise = np.random.normal(0,na,100)
    assetData = assetPure + assetNoise + (mf * benchNoise)
    sns.tsplot(assetData)      

    c = np.cov(np.array([assetData,benchData]))[0][1]
    beta = c / np.var(benchData)
    r_a = (assetData[-1] - assetData[0]) / assetData[0]
    alpha = r_a - (beta * r_b)
    alphas.append(v * alpha)
    
    
    print("This is endpoint: {}".format(endAsset))
    print("This is variance: {}".format(np.var(assetData)/np.var(benchData)))
    print("This is beta: {}".format(beta))
    print("This is r_b: {}".format(r_b))
    print("This is r_a: {}".format(r_a))
    print("This is alpha: {}".format(alpha))
    print("#######################################")

tot = 0
for v in range(100):
    tot += v
    makeAssetData(v)

print("This is av alpha: {}".format(np.sum(np.array(alphas))/v))
```

```python
def calcAlphaBetaInner(assetData, benchData):
    r_b = (benchData[-1] - benchData[0]) / benchData[0]
    r_a = (assetData[-1] - assetData[0]) / assetData[0]
    c = np.cov(np.array([assetData,benchData]))[0][1]
    beta = c / np.var(benchData)
    alpha = r_a - (beta * r_b)
    return(alpha, beta)
```

```python
def calcAlphaBeta(assetName, assetCol, benchName, benchCol, start=None, end=None):
    if end is None:
        end = datetime.today()
    if start is None:
        start = datetime.strptime("01 01 1900", '%m %d %Y')
    a = Quandl.get(assetName, returns="pandas")
    b = Quandl.get(benchName, returns="pandas")
    comb = pd.concat([a, b], axis="1").dropna()
    trunc = comb.ix[start:end]
    return (calcAlphaBetaInner(trunc[assetCol], trunc[benchCol]))
    

    
```
